```python
df.reset_index(inplace = True)
str_filter = df['Country'].str.contains('A')
df.loc[~str_filter, 'Country'] # 获得所有国家名字中不包含字母A的国家名
```

```python
str_filter2 = df['Country'].str.contains('A|Z')
df.loc[~str_filter2, 'Country'] # 获得所有国家名字中既不包含A也不包含Z的国家名
```

```python
str_filter2 = df['Country'].str.contains('A|B|C|D|E|F|G')
df.loc[~str_filter2, 'Country'] # 获得所有国家名字中不包含A到G的国家名
```

```python
str_filter3 = df['Country'].str.contains('[a-m]')
df.loc[~str_filter3, 'Country'] # 获得国家名中不包含小写字母a到m的国家名
```

```python
str_filter3 = df['Country'].str.contains('[A-Y]')
df.loc[~str_filter3, 'Country'] # 获得国家名中不包含大写字母A到 Y的国家名
```
