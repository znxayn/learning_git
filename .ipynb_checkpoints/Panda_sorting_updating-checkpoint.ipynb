```python
import pandas as pd
import numpy as np
csv_data_path = 'https://raw.githubusercontent.com/turingplanet/pandas-intro/main/public-datasets/small_survey_results.csv'
survey_df = pd.read_csv(csv_data_path)
```

```python
survey_df
```

```python
survey_df.set_index('Respondent', inplace = True) #将Respondent设为index
```

```python
survey_df
```

```python
survey_df.sort_index() #按照index进行排序，从小到大
```

```python
survey_df.sort_index(ascending = False) #按照index进行排序，从大到小
```

```python
survey_df.sort_values(by = 'Age', ascending=True)
```

```python
survey_df.sort_values(by=['Age', 'YearsCode'], ascending=False)
```

```python
survey_df.sort_values(by=['Age', 'YearsCode'], ascending=False)[['Age', 'YearsCode']]
```

```python
survey_df.sort_values(by=['Age', 'YearsCode'], ascending=[False, True]).head(10)[['Age', 'YearsCode']]
```

```python
survey_df.sort_values(by=['Age', 'YearsCode'], ascending=[False, False]).head(10)[['Age', 'YearsCode']]
```

```python
#如果需要改变原来的数据，加上 inplace=True
```

```python
survey_df['Age'].sort_values() # 将单列内容进行排序
```

```python
survey_df['ConvertedComp'].nlargest(10) # 前10名工资最高的用户工资信息
```

```python
survey_df['ConvertedComp'].nsmallest(10) # 工资最低的前十名用户工资信息
```

```python
richest_users = survey_df.nlargest(10, 'ConvertedComp')#获取最富有的10个人的信息
```

```python
richest_users
```

```python
richest_users[['ConvertedComp', 'DevType', 'EdLevel']] # 查看这些用户的工资、开发类型和教育程度
```

```python
poorest_users = survey_df.nsmallest(10, 'ConvertedComp') 
```

```python
poorest_users
```

```python
poorest_users[['ConvertedComp', 'DevType', 'EdLevel']] # 查看这些用户的工资、开发类型和教育程度
```
