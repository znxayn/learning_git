```python
csv_data_path = 'https://raw.githubusercontent.com/turingplanet/pandas-intro/main/public-datasets/country_info.csv'
```

```python
cou_info = pd.read_csv(csv_data_path)
cou_info
```

```python
cou_info.shape

```

```python
cou_info.head()
```

```python
cou_info.columns
```

```python
cou_info.iloc[100]
```

```python
cou_info.iloc[[2, 3], [0, 3]]
```

```python
cou_info.iloc[[2, 3], [3, 5]] #获取特定行及特定列
```

```python
cou_info.iloc[[1,2,3], [3,6,5]] #获取特定行及特定列并且可调整顺序，只能用数字索引，且不能连续行
```

```python
cou_info['Country'] = cou_info['Country'].str.strip() #把Country列中开头和尾部的空格去掉
```

```python
cou_info.set_index(['Country'], drop = True, inplace = True)
cou_info
```

```python
cou_info.loc[['China','Yemen'], ['Region']] #获取特定行及特定列并且可调整顺序，用loc即可用名称进行检索，使用innerlist
```

```python
cou_info.loc['China':'Japan', 'Region':'Net migration'] #可获取范围内的相关数据,此时用的是一个list
```
